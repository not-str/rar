```python
import numpy as np
from sklearn import datasets
x,y,coef=datasets.make_regression(n_samples=100,n_features=1,n_informative=1,noise=10,coef=True,random_state=0)  #features--> indepemdent var ;informtive-->dependent
x=np.interp(x,(x.min(),x.max()),(0,20)) #limiting them in range of 0-20
y=np.interp(y,(y.min(),y.max()),(20000,150000)) #limiting them in range of 20000-150000
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,'.',label="Training data")
plt.xlabel("Years Of experience")
plt.ylabel("Salary")
plt.title("Experince VS Salary") # this has got linear relation so move to next step
```

```python
#crreating regression model
from sklearn.linear_model import LinearRegression
reg_model=LinearRegression() #model created 
reg_model.fit(x,y) #fitting our varibles in model
y_predicted=reg_model.predict(x)
plt.plot(x,y_predicted,color="black")
plt.plot(x,y,'.',label="Training data")
plt.xlabel("Years Of experience")
plt.ylabel("Salary")
plt.title("Experince VS Salary")
```

```python
#extract this in dataframe
import pandas as pd
data={'Experience':np.round(x.flatten()),"Salary":np.round(y)}
df=pd.DataFrame(data)
df.head(10) #first 10 records
```
